```python
#日期轉換
def datechange(twdate):
    date=list(twdate)
    date.insert(0,twdate[3])
    date.insert(0,twdate[2])
    date.insert(0,twdate[1])
    date.insert(0,twdate[0])
    date.insert(0,"/")
    date.insert(0,twdate[5])
    date.insert(0,twdate[4])
    date.insert(0,"/")
    date.insert(0,twdate[7])
    date.insert(0,twdate[6])
    del(date[10:])#從第10開始刪除
    date="".join(date)
    return date
```
